```python
#group by Group column and data.percentage column, and get test/f1 column
grouped = results_csv.groupby(['Group', 'data.percentage']).agg({'test/f1': 'mean'}).reset_index()
```

```python
# get the rows where Group is baselines and data percentage is 1.0 , and the rows where group is patch_model
filtered = results_csv[((results_csv['Group'] == 'baseline') & (results_csv['data.percentage'] == 1.0)) | (results_csv['Group'] == 'patch_model')]
```

```python
#get test test/f1 column and group
test_f1 = filtered[['Group', 'data.percentage', 'test/f1','data.split']].sort_values(by=['Group', 'data.split'])
```

```python
test_f1
```

```python
from scipy import stats
f1s = test_f1[test_f1['Group'] == 'baseline']['test/f1'].values
compare_f1s = test_f1[test_f1['Group'] == 'patch_model']['test/f1'].values
t_stat, p_val = stats.ttest_ind(f1s, compare_f1s, equal_var=False, alternative='greater')
```
