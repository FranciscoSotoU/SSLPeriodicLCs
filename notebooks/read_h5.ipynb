```python
#map the classes to the new classes
new_labels = np.zeros(labels.shape)
for i in range(labels.shape[0]):
    new_labels[i] = map_classes[labels[i]]
```

```python
np.unique(new_labels, return_counts=True)
```

```python
test_partition = h5_file['test']
training_0_partition = h5_file['training_0']
training_1_partition = h5_file['training_1']
training_2_partition = h5_file['training_2']
training_3_partition = h5_file['training_3']
training_4_partition = h5_file['training_4']
validation_0_partition = h5_file['validation_0']
validation_1_partition = h5_file['validation_1']
validation_2_partition = h5_file['validation_2']
validation_3_partition = h5_file['validation_3']
validation_4_partition = h5_file['validation_4']
```

```python
new_important_classes = [0,1,2,3,4]
```

```python
new_indexes =np.where(np.isin(new_labels, new_important_classes))
```

```python
new_test_partition = test_partition[np.isin(test_partition, new_indexes)]
new_training_0_partition = training_0_partition[np.isin(training_0_partition, new_indexes)]
new_training_1_partition = training_1_partition[np.isin(training_1_partition, new_indexes)]
new_training_2_partition = training_2_partition[np.isin(training_2_partition, new_indexes)]
new_training_3_partition = training_3_partition[np.isin(training_3_partition, new_indexes)]
new_training_4_partition = training_4_partition[np.isin(training_4_partition, new_indexes)]
new_validation_0_partition = validation_0_partition[np.isin(validation_0_partition, new_indexes)]
new_validation_1_partition = validation_1_partition[np.isin(validation_1_partition, new_indexes)]
new_validation_2_partition = validation_2_partition[np.isin(validation_2_partition, new_indexes)]
new_validation_3_partition = validation_3_partition[np.isin(validation_3_partition, new_indexes)]
new_validation_4_partition = validation_4_partition[np.isin(validation_4_partition, new_indexes)]
```

```python
#add the new partitions to the h5 file
h5_file.create_dataset('test_reduced', data=new_test_partition)
h5_file.create_dataset('training_0_reduced', data=new_training_0_partition)
h5_file.create_dataset('training_1_reduced', data=new_training_1_partition)
h5_file.create_dataset('training_2_reduced', data=new_training_2_partition)
h5_file.create_dataset('training_3_reduced', data=new_training_3_partition)
h5_file.create_dataset('training_4_reduced', data=new_training_4_partition)
h5_file.create_dataset('validation_0_reduced', data=new_validation_0_partition)
h5_file.create_dataset('validation_1_reduced', data=new_validation_1_partition)
h5_file.create_dataset('validation_2_reduced', data=new_validation_2_partition)
h5_file.create_dataset('validation_3_reduced', data=new_validation_3_partition)
h5_file.create_dataset('validation_4_reduced', data=new_validation_4_partition)
h5_file.create_dataset('labels_reduced', data=new_labels)

```
