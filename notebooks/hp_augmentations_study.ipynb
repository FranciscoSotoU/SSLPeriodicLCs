```python
import pandas as pd
```

```python
vicreg_csv = pd.read_csv('/home/fsoto/Documents/LCsSSL/wandb_csv/hp_augmentations/single_aug_hp.csv')
```

```python
AUGS_DIRECTORY = "/home/fsoto/Documents/LCsSSL/wandb_csv/hp_augmentations"
```

## Amplitude Inverse


```python
amplitude_inverse_csv = pd.read_csv(f"{AUGS_DIRECTORY}/amplitude_inverse.csv") 
amplitude_inverse_hp = vicreg_csv[vicreg_csv['data.aug_names_list'] == '["amplitude_inverse"]'].sort_values(by='Created', ascending=True).reset_index(drop=True)
amplitude_inverse_hps= ['data.amplitude_inverse_prob']
amplitude_inverse_hp = amplitude_inverse_hp[amplitude_inverse_hps]
grouped_amplitude_inverse_f1 = amplitude_inverse_csv.groupby('model.pre_trained_iteration')['val/f1'].agg(['mean', 'std'])
grouped_amplitude_inverse_f1.join(amplitude_inverse_hp, on='model.pre_trained_iteration', how='left', lsuffix='_f1', rsuffix='_hp').sort_values(by='mean', ascending=False)
```

## Random Scaling


```python
random_scaling_csv = pd.read_csv(f"{AUGS_DIRECTORY}/random_scaling.csv")
random_scaling_hp = vicreg_csv[vicreg_csv['data.aug_names_list'] == '["random_scaling"]'].sort_values(by='Created', ascending=True).reset_index(drop=True)
random_scaling_hps = ['data.scaling_prob','data.scaling_sigma']
random_scaling_hp = random_scaling_hp[random_scaling_hps]
grouped_random_scaling_f1 = random_scaling_csv.groupby('model.pre_trained_iteration')['val/f1'].agg(['mean', 'std'])
grouped_random_scaling_f1.join(random_scaling_hp, on='model.pre_trained_iteration', how='left', lsuffix='_f1', rsuffix='_hp').sort_values(by='mean', ascending=False)
```

## Time Inverse

```python
time_inverse_csv = pd.read_csv(f"{AUGS_DIRECTORY}/time_inverse.csv")
time_inverse_hp = vicreg_csv[vicreg_csv['data.aug_names_list'] == '["time_inverse"]'].sort_values(by='Created', ascending=True).reset_index(drop=True)
time_inverse_hps = ['data.flip_time_signal_prob']
time_inverse_hp = time_inverse_hp[time_inverse_hps]
grouped_time_inverse_f1 = time_inverse_csv.groupby('model.pre_trained_iteration')['val/f1'].agg(['mean', 'std'])
grouped_time_inverse_f1.join(time_inverse_hp, on='model.pre_trained_iteration', how='left', lsuffix='_f1', rsuffix='_hp').sort_values(by='mean', ascending=False)
```

## Stretch

```python
stretch_csv = pd.read_csv(f"{AUGS_DIRECTORY}/stretch.csv")
stretch_hp = vicreg_csv[vicreg_csv['data.aug_names_list'] == '["stretch"]'].sort_values(by='Created', ascending=True).reset_index(drop=True)
stretch_hps = ['data.stretch_prob', 'data.stretch_sigma']
stretch_hp = stretch_hp[stretch_hps]
grouped_stretch_f1 = stretch_csv.groupby('model.pre_trained_iteration')['val/f1'].agg(['mean', 'std'])
grouped_stretch_f1.join(stretch_hp, on='model.pre_trained_iteration', how='left', lsuffix='_f1', rsuffix='_hp').sort_values(by='mean', ascending=False)
```

## Comparision

```python
# Get the best configuration for each augmentation
best_configs = []

# Amplitude Inverse
best_amplitude_inverse = grouped_amplitude_inverse_f1.join(amplitude_inverse_hp, on='model.pre_trained_iteration', how='left', lsuffix='_f1', rsuffix='_hp').sort_values(by='mean', ascending=False).iloc[0]
best_amplitude_inverse['augmentation'] = 'amplitude_inverse'
best_configs.append(best_amplitude_inverse)

# Random Scaling
best_random_scaling = grouped_random_scaling_f1.join(random_scaling_hp, on='model.pre_trained_iteration', how='left', lsuffix='_f1', rsuffix='_hp').sort_values(by='mean', ascending=False).iloc[0]
best_random_scaling['augmentation'] = 'random_scaling'
best_configs.append(best_random_scaling)

# Random Masking
best_random_masking = grouped_random_masking_f1.join(random_masking_hp, on='model.pre_trained_iteration', how='left', lsuffix='_f1', rsuffix='_hp').sort_values(by='mean', ascending=False).iloc[0]
best_random_masking['augmentation'] = 'random_masking'
best_configs.append(best_random_masking)

# Time Shift
best_time_shift = grouped_time_shift_f1.join(time_shift_hp, on='model.pre_trained_iteration', how='left', lsuffix='_f1', rsuffix='_hp').sort_values(by='mean', ascending=False).iloc[0]
best_time_shift['augmentation'] = 'time_shift'
best_configs.append(best_time_shift)

# Jitter
best_jitter = grouped_jitter_f1.join(jitter_hp, on='model.pre_trained_iteration', how='left', lsuffix='_f1', rsuffix='_hp').sort_values(by='mean', ascending=False).iloc[0]
best_jitter['augmentation'] = 'jitter'
best_configs.append(best_jitter)

# Channel Shuffle
best_channel_shuffle = grouped_channel_shuffle_f1.join(channel_shuffle_hp, on='model.pre_trained_iteration', how='left', lsuffix='_f1', rsuffix='_hp').sort_values(by='mean', ascending=False).iloc[0]
best_channel_shuffle['augmentation'] = 'channel_shuffle'
best_configs.append(best_channel_shuffle)

# Time Inverse
best_time_inverse = grouped_time_inverse_f1.join(time_inverse_hp, on='model.pre_trained_iteration', how='left', lsuffix='_f1', rsuffix='_hp').sort_values(by='mean', ascending=False).iloc[0]
best_time_inverse['augmentation'] = 'time_inverse'
best_configs.append(best_time_inverse)

# Signal Resampling
best_signal_resampling = grouped_signal_resampling_f1.join(signal_resampling_hp, on='model.pre_trained_iteration', how='left', lsuffix='_f1', rsuffix='_hp').sort_values(by='mean', ascending=False).iloc[0]
best_signal_resampling['augmentation'] = 'signal_resampling'
best_configs.append(best_signal_resampling)

# Stretch
best_stretch = grouped_stretch_f1.join(stretch_hp, on='model.pre_trained_iteration', how='left', lsuffix='_f1', rsuffix='_hp').sort_values(by='mean', ascending=False).iloc[0]
best_stretch['augmentation'] = 'stretch'
best_configs.append(best_stretch)

# Create the comparison dataframe
best_augmentations_df = pd.DataFrame(best_configs)

# Reorder columns to put augmentation first
cols = ['augmentation'] + [col for col in best_augmentations_df.columns if col != 'augmentation']
best_augmentations_df = best_augmentations_df[cols]

# Sort by mean F1 score (descending)
best_augmentations_df = best_augmentations_df.sort_values(by='mean', ascending=False).reset_index(drop=True)

print("Best configuration for each augmentation:")
best_augmentations_df
```
